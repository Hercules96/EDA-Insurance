### Feature Engineering


*   **Claim can be categories into low, avg, high**
*  **Bloodpressure can be categories.**


*  **bmi can be divided into multiple categories like underweight , obess , obesity,overweight**

```python
def cat_claim(claim_amt):
  if claim_amt<4800:
    return "Low claim"
  elif claim_amt>=4800 and claim_amt<16605:
    return "medium claim"
  elif claim_amt>=16605 and claim_amt<35000:
    return "high claim"
  else:
    return "very high claim"
```

```python
df['claim_cat'] = df['claim'].apply(cat_claim)
```

```python
df.head()
```

```python
df['bloodpressure'].describe()
```

```python
def cat_pressure(pressure):
  if pressure<120:
    return "Normal"
  elif pressure>= 120 and pressure<129:
    return "Elevated"
  elif pressure>=130 and pressure<=139:
    return "Stage 1-Hypertension"
  else :
    return "Stage 2-Hypertension"
```

```python
df['bloodpressure_cat'] = df['bloodpressure'].apply(cat_pressure)
```

```python
df.head()
```

```python
def cat_bmi(bmi):
  if bmi < 18.5:
    return "underweight"
  elif bmi>=18.5 and bmi<26:
    return "Normal"
  elif bmi>=26 and bmi<=30:
    return "Overweight"
  else :
    return "Obese"
```

```python
df['bmi_cat'] = df['bmi'].apply(cat_bmi)
```
